```python
import pandas as pd
```

```python
#To load data plese unpack Crimes-2016-2021 in data folder
Crimes_2016_2021 = pd.read_csv("data/Crimes-2016-2021.csv")
```

```python
Crimes_2016_2021['Date'] = pd.to_datetime(Crimes_2016_2021['Date'], format="%m/%d/%Y", exact=False)
Crimes_2016_2021['Updated On'] = pd.to_datetime(Crimes_2016_2021['Updated On'], format="%m/%d/%Y", exact=False)
```

```python
#Data regarding coordinates, FBI codes and community area will be discarded because it will not be a part of the scope.
#Spaces in column names are replayced by _.
Crimes = Crimes_2016_2021[['Case Number', 'Date', 'Updated On', 'Arrest', 'Domestic', 'Location Description', 
                           'IUCR', 'Primary Type', 'Description', 'District', 'Ward', 'Beat', 'Block']]
Crimes = Crimes.rename(columns = {'Location Description' : 'Location_Description', 'Primary Type' : 'Primary_Type', 
                                  'Case Number' : 'Case_Number', 'Updated On':'Updated_On'})
```

```python
#Some data is duplicated
Crimes.Case_Number.value_counts()
```

```python
Crimes[Crimes.Case_Number == "HZ140230"]
```

```python
#Duplicates are droped by Case_Number column because it is unique for every case
Crimes = Crimes.drop_duplicates(ignore_index = True, subset = 'Case_Number')
```

```python
Crimes.info()
```

```python
#Missing data is filled by Missing for Location Description and 0 for District and Ward
Crimes.Location_Description = Crimes.Location_Description.fillna('Missing')
Crimes.District = Crimes.District.fillna(0)
Crimes.Ward = Crimes.Ward.fillna(0)
Crimes = Crimes.astype({'District':'int64', 'Ward':'int64'})
```

```python
Crimes.info()
```

```python
#New dataframe is created, location, which will be a second table in postgresql
location = Crimes[['Block', 'District', 'Ward', 'Beat']]
location = location.drop_duplicates(ignore_index = True)
```

```python
#Location doesn't have column with unique values, it requres a special column - location_id containing index numbers 
#to create relation
location['location_id'] = location.index + 1
location
```

```python
#2 new csv files are created in data/ directory
location.to_csv('data/location.csv', index = False)
Crimes.to_csv('data/Crimes.csv', index = False)
```
